```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df= pd.read_csv('../data/togo-dapaong_qc.csv')

# Display the first few rows of the dataset
df.head()
```

```python
df.describe()
```

```python
# Missing value count
missing_counts = df.isna().sum()
print("Missing values:\n", missing_counts)

# Columns with >5% nulls
null_threshold = 0.05 * len(df)
high_nulls = missing_counts[missing_counts > null_threshold]
print("\nColumns with >5% nulls:\n", high_nulls)
```

```python
from scipy.stats import zscore
import numpy as np

# Columns of interest
cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores
z_scores = df[cols].apply(zscore)
outliers = (np.abs(z_scores) > 3)
outlier_rows = df[outliers.any(axis=1)]
print(f"\nOutliers detected: {outlier_rows.shape[0]} rows")

# Optionally drop or impute missing values
df_cleaned = df.copy()
df_cleaned[cols] = df_cleaned[cols].fillna(df_cleaned[cols].median())
```

```python
import os

# Ensure the folder exists
os.makedirs("../cleaned_data", exist_ok=True)

# Save the cleaned DataFrame with a standardized filename
df_cleaned.to_csv("../cleaned_data/cleaned_togo_data.csv", index=False)

print("✅ Cleaned data saved to ../cleaned_data/cleaned_togo_data.csv")
```

```python
#  Time Series Analysis
import matplotlib.pyplot as plt

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df.set_index('Timestamp')[['GHI', 'DNI', 'DHI', 'Tamb']].plot(figsize=(14, 6))
plt.title("GHI, DNI, DHI, Tamb Over Time")
plt.ylabel("Value")
plt.show()
```

```python
# Cleaning Impact on ModA & ModB
if 'Cleaning' in df.columns:
    df.groupby('Cleaning')[['ModA', 'ModB']].mean().plot(kind='bar', figsize=(8, 4))
    plt.title("ModA & ModB Averages by Cleaning Flag")
    plt.ylabel("Sensor Reading")
    plt.show()
```

```python
# correlation and scatter analysis
import seaborn as sns

# Heatmap
corr_cols = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
sns.heatmap(df[corr_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Scatter plots
sns.scatterplot(x='WS', y='GHI', data=df)
plt.title("Wind Speed vs. GHI")
plt.show()

sns.scatterplot(x='RH', y='Tamb', data=df)
plt.title("Relative Humidity vs. Temperature")
plt.show()
```

```python
# wind and distribution analysis

# Histogram
df['GHI'].hist(bins=30)
plt.title("Histogram of GHI")
plt.xlabel("GHI")
plt.show()

df['WS'].hist(bins=30)
plt.title("Histogram of Wind Speed")
plt.xlabel("WS")
plt.show()

```

```python
# Optional Wind Rose Plot
from windrose import WindroseAxes

ax = WindroseAxes.from_ax()
ax.bar(df['WD'], df['WS'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
plt.title("Wind Rose")
plt.show()
```

```python
# temperature and RH Influence
sns.scatterplot(x='RH', y='GHI', data=df)
plt.title("Relative Humidity vs. GHI")
plt.show()
```

```python
# Bubble Chart: GHI vs Tamb, bubble size = RH
plt.figure(figsize=(10, 6))
plt.scatter(df['Tamb'], df['GHI'], s=df['RH'], alpha=0.5)
plt.xlabel('Tamb')
plt.ylabel('GHI')
plt.title('Bubble Chart: GHI vs. Tamb (bubble size = RH)')
plt.show()
```
